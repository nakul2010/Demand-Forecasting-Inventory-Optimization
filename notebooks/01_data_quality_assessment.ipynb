```python
df["date"] = pd.to_datetime(df["date"])

print("Start Date:", df["date"].min())
print("End Date:", df["date"].max())
```

```python
sales_by_store = df.groupby("store")["sales"].sum().sort_values(ascending=False)

print(sales_by_store)
```
