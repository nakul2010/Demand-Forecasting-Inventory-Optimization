```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/train.csv")

df["date"] = pd.to_datetime(df["date"])

df.head()
```

```python
df = df.sort_values(
    by=["store", "item", "date"]
)

df.head()
```

```python
df["year"] = df["date"].dt.year

df["month"] = df["date"].dt.month

df["quarter"] = df["date"].dt.quarter

df["day"] = df["date"].dt.day

df["day_of_week"] = df["date"].dt.dayofweek

df["day_of_year"] = df["date"].dt.dayofyear

df["week_of_year"] = df["date"].dt.isocalendar().week.astype(int)

df.head()
```

```python
df["is_weekend"] = np.where(
    df["day_of_week"] >= 5,
    1,
    0
)

df[["date", "day_of_week", "is_weekend"]].head()
```

```python
df["lag_7"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .shift(7)
)
```

```python
df["lag_14"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .shift(14)
)
```

```python
df["lag_30"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .shift(30)
)
```

```python
df["lag_90"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .shift(90)
)
```

```python
df[
    [
        "date",
        "store",
        "item",
        "sales",
        "lag_7",
        "lag_14"
    ]
].head(20)
```

```python
df["rolling_mean_7"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .transform(
        lambda x: x.shift(1).rolling(7).mean()
    )
)
```

```python
df["rolling_mean_30"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .transform(
        lambda x: x.shift(1).rolling(30).mean()
    )
)
```

```python
df["rolling_std_7"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .transform(
        lambda x: x.shift(1).rolling(7).std()
    )
)
```

```python
df["rolling_std_30"] = (
    df.groupby(
        ["store", "item"]
    )["sales"]
    .transform(
        lambda x: x.shift(1).rolling(30).std()
    )
)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()

df.shape
```
